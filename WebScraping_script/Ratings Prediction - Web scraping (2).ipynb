## 1.1 Scraping review rating for ratchet screwdriver set

```python
# Listing out url for calling Function
function=[]
for i in range(1,501):
    c_f=(f"https://www.flipkart.com/jackly-square-precision-32-pc-ratchet-screwdriver-set/product-reviews/itmee8fwhfecezhz?pid=SCSEE8FWZQMBQKQU&lid=LSTSCSEE8FWZQMBQKQU0FINF6&marketplace=FLIPKART&page={i}")
    function.append(c_f)
```

```python
# Function Definition
review=[]
rating=[]
def tool_kit(url):
    driver=webdriver.Chrome(r"c://chromedriver.exe")# Activating the chrome browser
    driver.get(url)
    r=requests.get(url)
    soup=BeautifulSoup(r.content,'html.parser')
    for i in soup.find_all("div",class_="t-ZTKy"):
        review.append(i.text)
    for i in soup.find_all("div",class_=["_3LWZlK _1BLPMq","_3LWZlK _32lA32 _1BLPMq","_3LWZlK _1rdVr6 _1BLPMq"]):
        rating.append(i.text)
                
# Calling Function
for i in function:
    tool_kit(i) 
```

## 1.4 Scraping review rating for Fire TV Stick

```python
# Listing out url for calling Function
function=[]
for i in range(1,501):
    c_f=(f"https://www.amazon.in/Bourge-Vega-Pearl-z2-Running-Shoes/product-reviews/B08VXPYDM9/ref=cm_cr_arp_d_paging_btm_next_2?ie=UTF8&reviewerType=all_reviews&pageNumber={i}")
    function.append(c_f)
```

```python
function
```

```python
# Function Definition
# Function Definition
review_list=[]
def running_shoe(url):
    driver=webdriver.Chrome(r"c://chromedriver.exe")# Activating the chrome browser
    driver.get(url)
    r=requests.get(url)
    soup=BeautifulSoup(r.content,'html.parser')
    reviews=soup.find_all('div',{'data-hook' : 'review'})
    try:
        for item in reviews:
            review={    
            'Review' : item.find('span',{'data-hook':'review-body'}).text.strip(),
            'Rating': float(item.find('i',{'data-hook':'review-star-rating'}).text.replace('out of 5 stars',"").strip())
            }
            review_list.append(review)
    except:
        pass   
#calling function
for i in function:
    running_shoe(i) 
```

```python
df = pd.DataFrame(review_list)
df.to_csv('RR4', index=False)
df
```
